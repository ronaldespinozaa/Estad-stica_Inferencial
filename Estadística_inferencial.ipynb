# Estadística inferencial

## Conceptos previos:

- Explicar la diferencia entre estadística descriptiva e inferencial.
La estadística inferencial es una rama de la estadística que se centra en hacer inferencias o generalizaciones sobre una población basándose en una muestra de datos de esa población. En otras palabras, nos ayuda a sacar conclusiones sobre un grupo más grande utilizando información de un conjunto más pequeño pero representativo de ese grupo.



- Recordar los conceptos de población y muestra.  
    Una muestra es un subconjunto representativo de la población. Dado que estudiar toda la población puede ser impráctico o costoso, se selecciona una muestra para realizar observaciones y análisis. Según la prueba que se quiere realizar, los datos disponibles y otras variables podemos elegir entre diferentes tipos de muestreo(Muestreo aleatorio simple, muestreo estratificado, muestreo por Conglomerados,muestreo sistemático, etc ) 


![población_muestra](./poblacion_muestra.jpeg)


- Diferenciar parámetros y estadísticos  
    Un parámetro es una medida numérica que describe alguna característica de una población. Por ejemplo, la media (μ) y la desviación estándar (σ) son parámetros poblacionales.ç
    
    Un estadístico es una medida numérica que describe alguna característica de una muestra. Por ejemplo, la media muestral y la desviación estándar muestral son estadísticos.

    Entonces, el parámetro se refiere a características poblacionales, mientras que el estadístico se refiere a características de muestras tomadas de esa población. Es importante entender esta distinción al realizar inferencias estadísticas basadas en datos muestrales.

## Hipótesis Estadística:  
Una hipótesis estadística es una afirmación o suposición acerca de la población que se está probando. Puede ser una afirmación sobre un parámetro específico, como la media.

## Pruebas de Hipótesis:  
Las pruebas de hipótesis son procedimientos estadísticos que nos permiten tomar decisiones acerca de las afirmaciones hechas sobre una población basándonos en la información contenida en una muestra. Esto implica comparar evidencia empírica con una afirmación dada.

## Hipótesis nula

Las hipótesis nulas son afirmaciones que se hacen sobre una población o un fenómeno, y generalmente se expresan como la ausencia de un efecto o relación.Es importante recordar que la hipótesis nula generalmente representa la idea de que no hay efecto, relación o diferencia significativa en la población subyacente.

## Hipótesis alternativa

Las hipótesis alternativas expresan la idea de que hay algún efecto, relación o diferencia significativa en la población. se formulan para expresar la dirección o naturaleza del efecto que se está investigando en el contexto de diferentes pruebas estadísticas. La elección de la hipótesis alternativa depende de la pregunta de investigación y de lo que el investigador está tratando de demostrar.


## Nivel de significancia (α)

El nivel de significancia (α) es la probabilidad de cometer un error tipo I en una prueba estadística. Es la probabilidad de rechazar incorrectamente la hipótesis nula cuando es verdadera. Comúnmente, se establece el nivel de significancia antes de realizar una prueba y se denota por un valor como 0.05 o 0.01.
 
## Errores Tipo I y Tipo II.

### Error Tipo I (Falso Positivo):
- Definición: Ocurre cuando se rechaza incorrectamente la hipótesis nula cuando es verdadera.

- Probabilidad de Error Tipo I (α): Es igual al nivel de significancia establecido antes de realizar la prueba.

- Consecuencia: A menudo se considera más grave en el contexto de la investigación, ya que implica afirmar falsamente la existencia de un efecto o diferencia cuando no hay evidencia suficiente para hacerlo.

### Error Tipo II (Falso Negativo):

- Definición: Ocurre cuando no se rechaza la hipótesis nula cuando es falsa.

- Probabilidad de Error Tipo II (β): Representa la probabilidad de no rechazar la hipótesis nula cuando la hipótesis alternativa es verdadera.

- Consecuencia: Indica que no se ha identificado un efecto o diferencia real cuando existe.


## Intervalo de Confianza:
Un intervalo de confianza (IC) es un rango de valores que se utiliza para estimar un parámetro desconocido de una población. La idea principal detrás de un intervalo de confianza es proporcionar una estimación de la variabilidad asociada con la estimación puntual del parámetro. En lugar de dar un único número como estimación, se proporciona un rango dentro del cual se espera que se encuentre el valor verdadero del parámetro con cierto nivel de confianza.

### Nivel de Confianza (1 - α):

El nivel de confianza representa la probabilidad de que el intervalo de confianza capture el verdadero valor del parámetro.
Comúnmente usado: 95% de confianza (α=0.05) o 99% de confianza (α=0.01).
Ejemplo:

Un intervalo de confianza del 95% para la media de una población (digamos, la media de las alturas de una población) podría ser de 165 cm a 175 cm.
Esto se interpretaría como: "Estamos 95% seguros de que la verdadera media de la altura de la población está entre 165 cm y 175 cm".


```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Generar datos de ejemplo (puedes reemplazar esto con tus propios datos)
np.random.seed(42)
datos = np.random.normal(loc=20, scale=5, size=100)

# Calcular el intervalo de confianza para la media con un nivel de confianza del 95%
intervalo_confianza = stats.norm.interval(0.95, loc=np.mean(datos), scale=stats.sem(datos))

# Crear un gráfico KDE de los datos
sns.kdeplot(datos, shade=True, color='blue', label='Densidad de probabilidad')

# Agregar líneas para mostrar la media y el intervalo de confianza
media = np.mean(datos)
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label='Media de la muestra')
plt.axvline(intervalo_confianza[0], color='green', linestyle='dashed', linewidth=2, label='Intervalo de confianza')
plt.axvline(intervalo_confianza[1], color='green', linestyle='dashed', linewidth=2)

# Agregar etiquetas y leyenda
plt.xlabel('Valores')
plt.ylabel('Densidad de probabilidad')
plt.title('Gráfico KDE y Intervalo de Confianza')
plt.legend()

# Mostrar el gráfico
plt.show()
```


## 2. Pruebas No Paramétricas:
No hacen suposiciones específicas sobre la distribución de los datos o asumen una distribución más general. 
 


```python
from scipy.stats import mannwhitneyu, kruskal, friedmanchisquare, chi2_contingency, ks_2samp, chi2, anderson, pearsonr, linregress

# Configuración de semilla para reproducibilidad
np.random.seed(42)

# Crear un DataFrame de ejemplo
data = {'Grupo_1': np.random.choice(['A', 'B', 'C'], size=50),
        'Grupo_2': np.random.choice(['A', 'B', 'C'], size=50),
        'Preferencia_1': np.random.choice(['Dulce', 'Salado', 'Ácido'], size=50),
        'Preferencia_2': np.random.choice(['Dulce', 'Salado', 'Ácido'], size=50),
        'Calificación': np.random.randint(1, 11, size=50)}

df_nonparametric = pd.DataFrame(data)
df_nonparametric
```

Estamos estudiando la preferencia de sabor de tres tipos diferentes de bebidas (A, B y C) entre dos grupos de consumidores (Grupo_1 y Grupo_2)

a. Pruebas para Dos Muestras Independientes:
- Prueba de Mann-Whitney (U de Mann-Whitney).  


```python
# Prueba de Mann-Whitney para dos muestras independientes
stat_mw, p_value_mw = mannwhitneyu(df_nonparametric['Calificación'][df_nonparametric['Grupo_1'] == 'A'],
                                    df_nonparametric['Calificación'][df_nonparametric['Grupo_1'] == 'B'])

# Hipótesis nula (H0): No hay diferencia significativa en la calificación entre los grupos 1A y 1B.
# Hipótesis alternativa (Ha): Existe una diferencia significativa en la calificación entre los grupos 1A y 1B.

alpha = 0.05  # Nivel de significancia

print(f"\nPrueba de Mann-Whitney para dos muestras independientes: stat = {stat_mw}, p_value = {p_value_mw}")

if p_value_mw < alpha:
    print("Rechazamos la hipótesis nula. Hay evidencia de una diferencia significativa en la calificación entre los grupos 1A y 1B.")
else:
    print("No hay suficiente evidencia para rechazar la hipótesis nula. No hay diferencia significativa en la calificación entre los grupos 1A y 1B.")
```



b. Pruebas para Más de Dos Muestras Independientes:
- Prueba de Kruskal-Wallis.  



```python
# Prueba de Kruskal-Wallis para más de dos muestras independientes
stat_kw, p_value_kw = kruskal(df_nonparametric['Calificación'][df_nonparametric['Grupo_1'] == 'A'],
                               df_nonparametric['Calificación'][df_nonparametric['Grupo_1'] == 'B'],
                               df_nonparametric['Calificación'][df_nonparametric['Grupo_1'] == 'C'])

# Hipótesis nula (H0): No hay diferencia significativa en la calificación entre los grupos 1A, 1B y 1C.
# Hipótesis alternativa (Ha): Existe al menos una diferencia significativa en la calificación entre los grupos 1A, 1B y 1C.

print(f"\nPrueba de Kruskal-Wallis para más de dos muestras independientes: stat = {stat_kw}, p_value = {p_value_kw}")

if p_value_kw < alpha:
    print("Rechazamos la hipótesis nula. Hay evidencia de al menos una diferencia significativa en la calificación entre los grupos 1A, 1B y 1C.")
else:
    print("No hay suficiente evidencia para rechazar la hipótesis nula. No hay diferencia significativa en la calificación entre los grupos 1A, 1B y 1C.")
```

- Prueba de Friedman para muestras relacionadas (análoga a ANOVA no paramétrica).   


```python

# Asegurarnos de que todas las muestras tengan el mismo número de observaciones
min_observation_count = min(df_nonparametric.groupby(['Grupo_1', 'Preferencia_1']).size().min(),
                            df_nonparametric.groupby(['Grupo_2', 'Preferencia_2']).size().min())

# Función para ajustar el número de observaciones por grupo y preferencia
def adjust_observation_count(group):
    return group.sample(min_observation_count)

# Aplicar la función a cada grupo y preferencia
df_nonparametric_adjusted = df_nonparametric.groupby(['Grupo_1', 'Preferencia_1']).apply(adjust_observation_count)
df_nonparametric_adjusted.reset_index(drop=True, inplace=True)

# Prueba de Friedman para más de dos muestras emparejadas
stat_friedman, p_value_friedman = friedmanchisquare(df_nonparametric_adjusted['Calificación'][df_nonparametric_adjusted['Preferencia_1'] == 'Dulce'],
                                                    df_nonparametric_adjusted['Calificación'][df_nonparametric_adjusted['Preferencia_1'] == 'Salado'],
                                                    df_nonparametric_adjusted['Calificación'][df_nonparametric_adjusted['Preferencia_1'] == 'Ácido'])

print(f"Prueba de Friedman para más de dos muestras emparejadas (con igual número de observaciones): stat = {stat_friedman}, p_value = {p_value_friedman}")

if p_value_friedman < alpha:
    print("Rechazamos la hipótesis nula. Hay evidencia de una asociación significativa entre el grupo y la preferencia de sabor.")
else:
    print("No hay suficiente evidencia para rechazar la hipótesis nula. No hay asociación significativa entre el grupo y la preferencia de sabor.")
```

```python
# Prueba de Chi cuadrado de independencia
stat_chi2_independence, p_value_chi2_independence, _, _ = chi2_contingency(pd.crosstab(df_nonparametric['Grupo_1'], df_nonparametric['Preferencia_1']))

tabla_contingencia_chi = pd.crosstab(df_nonparametric['Grupo_1'], df_nonparametric['Preferencia_1'])
tabla_contingencia_chi.index = ['A', 'B','C']
print(tabla_contingencia_chi)

# Hipótesis nula (H0): No hay asociación significativa entre el grupo y la preferencia de sabor.
# Hipótesis alternativa (Ha): Existe una asociación significativa entre el grupo y la preferencia de sabor.

print(f"\nPrueba de Chi cuadrado de independencia: stat = {stat_chi2_independence}, p_value = {p_value_chi2_independence}")

if p_value_chi2_independence < alpha:
    print("Rechazamos la hipótesis nula. Hay evidencia de una asociación significativa entre el grupo y la preferencia de sabor.")
else:
    print("No hay suficiente evidencia para rechazar la hipótesis nula. No hay asociación significativa entre el grupo y la preferencia de sabor.")
```
